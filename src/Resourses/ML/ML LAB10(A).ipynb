```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/TANISHKA DODIYA/Desktop/SEMESTER 7/ML/Income.csv')
df
```

```python
plt.scatter(df['Age'], df['Income'])
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income'],color='green')
plt.scatter(df2.Age,df2['Income'],color='red')
plt.scatter(df3.Age,df3['Income'],color='black')
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income']])
df['Income']=scaler.transform(df[['Income']])
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income'],color='green')
plt.scatter(df2.Age,df2['Income'],color='red')
plt.scatter(df3.Age,df3['Income'],color='black')
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income'],color='green')
plt.scatter(df2.Age,df2['Income'],color='red')
plt.scatter(df3.Age,df3['Income'],color='black')
plt.xlabel('Age')
plt.ylabel('Income')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',label='centroid')
```
